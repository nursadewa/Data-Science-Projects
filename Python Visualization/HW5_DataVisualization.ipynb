### **Soal No.1:** Buatlah 5 visualisasi masing-masing untuk kolom attrition_flag, customer_age, gender, education_level, dan income_category. (Wajib ada pie chart, donut chart, vertical bar chart, dan horizontal bar chart)

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set plot style
sns.set(style="whitegrid")

# Create subplots for each visualization type
fig, axs = plt.subplots(5, 4, figsize=(20, 25))

# List of columns to visualize
columns = ['Attrition_Flag', 'Customer_Age', 'Gender', 'Education_Level', 'Income_Category']

for i, column in enumerate(columns):
    # Pie Chart
    data = cc_data[column].value_counts()
    axs[i, 0].pie(data, labels=data.index, autopct='%1.1f%%', startangle=140)
    axs[i, 0].set_title(f'{column} - Pie Chart')

    # Donut Chart
    wedges, texts, autotexts = axs[i, 1].pie(data, labels=data.index, autopct='%1.1f%%', startangle=140, wedgeprops=dict(width=0.3))
    axs[i, 1].set_title(f'{column} - Donut Chart')

    # Vertical Bar Chart
    sns.countplot(y=column, data=cc_data, ax=axs[i, 2])
    axs[i, 2].set_title(f'{column} - Vertical Bar Chart')

    # Horizontal Bar Chart
    sns.countplot(x=column, data=cc_data, ax=axs[i, 3])
    axs[i, 3].set_title(f'{column} - Horizontal Bar Chart')

# Adjust layout to prevent overlap
plt.tight_layout()
plt.show()
```

## **Soal No.5:** Buatlah 1 visualisasi menggunakan heatmap yang memadukan lebih dari satu kolom (bebas antar kolom yang mana saja)

```python
# Memilih kolom yang memiliki korelasi dan tipe data sama
selected_columns = ['Customer_Age', 'Dependent_count', 'Months_on_book', 'Credit_Limit',
                    'Total_Revolving_Bal', 'Total_Trans_Amt', 'Total_Trans_Ct']
correlation = cc_data[selected_columns].corr()

# Set up the figure size
plt.figure(figsize=(10, 8))

# Create a heatmap using the correlation matrix
plt.rcParams.update({'font.size': 12})
sns.heatmap(correlation, cmap = 'viridis', vmin = -1, vmax = 1, center = 0, annot=True, fmt=".2f", square=True, linewidths=.5)

# Add title
plt.title('Heatmap of Correlation for Selected Columns')

# Show the plot
plt.show()
```
