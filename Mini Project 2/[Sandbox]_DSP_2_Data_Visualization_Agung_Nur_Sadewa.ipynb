## Columns Renaming

```python
rename_mapper = {col:col.replace(" ", "_").lower() for col in df.columns} #dict comprehension
df.rename(
    columns=rename_mapper,
    inplace=True
)

df.head()
```

## To Datetime datatype

```python
df["order_date"] = pd.to_datetime(df["order_date"])
df["ship_date"] = pd.to_datetime(df["ship_date"])
```

```python
df[["order_date", "ship_date"]].head().info()
```

```python
df = df.sort_values(
    by=["order_id", "order_date", "ship_date"],
    ascending=[True, True, True],
    ignore_index=True
)

df.head(20)
```

## Remove "$" in ["sales", "profit", "shipping_cost" ]

```python
dollar_cols = ["sales", "profit", "shipping_cost"]

for col in dollar_cols:
  df[col] = (
      df[col]
      .str.replace("[^\d.]","", regex =True)
  )
  df.rename(columns={col: col+"($)"}, inplace=True)

print("Formatting & Renaming, Done!")
```

## Monthly Sales Trends

```python
monthly_sum_sales = df.groupby("months").agg({"sales($)": "sum"})
monthly_sum_sales
```

```python
ordered_months = [
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
]

Month_CI = pd.CategoricalIndex(
    monthly_sum_sales.index,
    ordered=True,
    categories=ordered_months
)

Month_CI
```

```python
monthly_sum_sales.index = Month_CI
monthly_sum_sales = monthly_sum_sales.sort_index()
monthly_sum_sales
```

```python
import matplotlib.ticker as ticker
```

```python
# Create line chart for monthly sales
fig, ax = plt.subplots(figsize=(10, 5))

ax.plot(
    monthly_sum_sales.index,
    monthly_sum_sales["sales($)"],
    marker='*',
    mfc='red', # marker face color
    mec='red', # marker edge color
    linestyle='-'
)

ax.set_title('Monthly Sales Trends')
ax.set_xlabel('Month')
ax.set_ylabel('Sales')
ax.grid(True)
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, pos: '$' + '{:,.2f}'.format(x/1000) + 'K'))

# Show the line chart
fig.show()
```

## 5.&nbsp; **Customer-wise Analysis: Top 10 Highest Customers Sales**

```python
customer_order = df.groupby("customer_name").agg({"sales($)": "sum"}).sort_values(by='sales($)', ascending=False)
customer_order
```

```python
top_10_customers = customer_order.nlargest(10, 'sales($)')
top_10_customers
```

```python
# Identify the indices of the highest and lowest values
max_index = top_10_customers['sales($)'].idxmax()
min_index = top_10_customers['sales($)'].idxmin()

# Create explode list, where the highest and lowest values are offset
explode = [0.1 if customer == max_index or customer == min_index else 0 for customer in top_10_customers.index]

# Create a pie chart with exploded wedges, shadow, and custom wedge properties
plt.figure(figsize=(10, 7))
wedges, texts, autotexts = plt.pie(
    top_10_customers['sales($)'],
    labels=top_10_customers.index,
    autopct='%1.1f%%',
    startangle=140,
    shadow=True,  # Add shadow
    explode=explode,  # Explode the highest and lowest wedges
    wedgeprops={'edgecolor': 'black', 'linewidth': 3}  # Customize wedge properties
)

# Customize the text appearance
for text in texts + autotexts:
    text.set_fontsize(10)
    text.set_color('black')

plt.title('Top 10 Customers by Sales($)')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```

```python
# Define a list of colors for the bars
colors = ['red', 'blue', 'green', 'purple', 'orange', 'brown', 'pink', 'gray', 'cyan', 'magenta']

# Plotting the bar chart
plt.figure(figsize=(10, 7))
bars = plt.bar(top_10_customers.index, top_10_customers['sales($)'], color= colors)

# Adding labels and title
plt.xlabel('Customer Name')
plt.ylabel('Sales ($)')
plt.title('Top 10 Customers by Sales($)')
plt.xticks(rotation=45)
plt.tight_layout()

# Show plot
plt.show()
```
